```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

#Load Data
df = pd.read_csv("road_accident_statistics.csv")

#View Basic Dataset Info
print("ðŸ”¹ Dataset Shape:", df.shape)
print("\nðŸ”¹ Columns:", df.columns.tolist())
print("\nðŸ”¹ Missing Values:\n", df.isnull().sum())
print("\nðŸ”¹ Data Types:\n", df.dtypes)

#Summary Statistics
print("\nðŸ”¹ Summary Statistics:\n", df.describe())

#Correlation Matrix
print("\nðŸ”¹ Correlation Matrix:\n", df.corr(numeric_only=True))
```

```python
# Line plot for total accidents and deaths
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Number of Accidents'], label='Accidents', linewidth=2)
plt.plot(df['Year'], df['Death'], label='Deaths', linewidth=2, linestyle='--')
plt.title('Trend of Accidents and Deaths Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Line plot for all injury types
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Number of Serious Injuries'], label='Serious Injuries')
plt.plot(df['Year'], df['Number of Minor Injuries'], label='Minor Injuries')
plt.plot(df['Year'], df['Number of Moderate Injuries'], label='Moderate Injuries')
plt.plot(df['Year'], df['Number of Severe Injuries'], label='Severe Injuries')
plt.title('Injury Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Injury Count')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Prepare data for stacked bar chart
injury_df = df[['Year', 'Number of Serious Injuries', 'Number of Minor Injuries',
                'Number of Moderate Injuries', 'Number of Severe Injuries']]
injury_df.set_index('Year', inplace=True)

# Plot stacked bar chart
injury_df.plot(kind='bar', stacked=True, figsize=(14, 7))
plt.title('Year-wise Injury Severity Distribution')
plt.xlabel('Year')
plt.ylabel('Total Injuries')
plt.tight_layout()
plt.show()
```

```python
# Find year with most accidents and deaths
max_accidents = df[df['Number of Accidents'] == df['Number of Accidents'].max()]
min_accidents = df[df['Number of Accidents'] == df['Number of Accidents'].min()]

max_deaths = df[df['Death'] == df['Death'].max()]
min_deaths = df[df['Death'] == df['Death'].min()]

print("\nðŸ”º Year with Maximum Accidents:\n", max_accidents)
print("\nðŸ”» Year with Minimum Accidents:\n", min_accidents)
print("\nðŸ’€ Year with Maximum Deaths:\n", max_deaths)
print("\nðŸ˜‡ Year with Minimum Deaths:\n", min_deaths)
```

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

#Load your dataset
df = pd.read_csv("road_accident_statistics.csv")

#Select features and target
X = df[['Year']]  # Feature
y = df['Number of Accidents']  # Target

#Train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

#Predict on training data
y_pred = model.predict(X)

#Evaluate the model
print("ðŸ“‰ Mean Squared Error:", mean_squared_error(y, y_pred))
print("ðŸ“ˆ RÂ² Score:", r2_score(y, y_pred))

#Plot actual vs predicted
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Predicted Trend')
plt.title("Accident Prediction using Linear Regression")
plt.xlabel("Year")
plt.ylabel("Number of Accidents")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Predict for future years (2024 to 2030)
future_years = pd.DataFrame({'Year': list(range(2024, 2031))})
future_predictions = model.predict(future_years)

# Display future predictions
print("\nðŸ”® Predicted Accidents for Future Years:")
for year, prediction in zip(future_years['Year'], future_predictions):
    print(f"{year}: {int(prediction)} accidents")
```
